#Lightgbm Hyperparameter

```python

from lightgbm import LGBMRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import numpy as np

# Parameter grid for tuning
param_dist = {
    "num_leaves": [20, 31, 40, 50, 60],
    "max_depth": [-1, 5, 7, 9, 12],
    "learning_rate": [0.01, 0.03, 0.05, 0.1],
    "n_estimators": [300, 500, 700, 900],
    "subsample": [0.6, 0.7, 0.8, 0.9, 1.0],
    "colsample_bytree": [0.6, 0.7, 0.8, 0.9, 1.0],
    "reg_alpha": [0, 0.01, 0.1, 0.5],
    "reg_lambda": [0, 0.01, 0.1, 0.5]
}

# Base model
lgbm_tune = LGBMRegressor(random_state=42)

# Randomized Search
lgbm_search = RandomizedSearchCV(
    estimator=lgbm_tune,
    param_distributions=param_dist,
    n_iter=30,             # number of combinations to try
    scoring='r2',
    cv=3,
    verbose=1,
    random_state=42,
    n_jobs=-1
)

# Fit search
lgbm_search.fit(X_train, y_train)

print("\nBest Hyperparameters:")
print(lgbm_search.best_params_)

# Get best tuned model
best_lgbm = lgbm_search.best_estimator_

# Predict
y_pred_best_lgbm = best_lgbm.predict(X_test)

# Evaluation
r2 = r2_score(y_test, y_pred_best_lgbm)
mse_lgbm = mean_squared_error(y_test, y_pred_best_lgbm)

mae = mean_absolute_error(y_test, y_pred_best_lgbm)
rmse = mean_squared_error(y_test, y_pred_best_lgbm)**0.5

print("\n======= Tuned LightGBM Results =======")
print("R2 Score:", r2)
print("MAE:", mae)
print("MSE (LightGBM):", mse_lgbm)
print("RMSE:", rmse)
```

```python
plt.figure(figsize=(12,6))

plt.plot(sorted(y_test.values), label="Actual (sorted)", linewidth=2)
plt.plot(sorted(y_pred_best_lgbm), label="Predicted (sorted)", linewidth=2)

plt.title("Actual vs Predicted (Sorted Line Plot)")
plt.xlabel("Sample Index")
plt.ylabel("Power Generated")
plt.legend()
plt.show()
```

```python
from catboost import CatBoostRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

cat = CatBoostRegressor(
    loss_function="RMSE",
    random_state=42,
    verbose=0
)

param_grid = {
    "iterations": [200, 400, 600],
    "learning_rate": [0.01, 0.05, 0.1],
    "depth": [4, 6, 8],
    "l2_leaf_reg": [1, 3, 5]
}

# 3. GridSearchCV

grid = GridSearchCV(
    estimator=cat,
    param_grid=param_grid,
    scoring="r2",
    cv=5,
    n_jobs=-1,
    verbose=2
)

grid.fit(X_train, y_train)

# 4. Best Params

print("Best CatBoost Params:", grid.best_params_)

# 5. Final Model Prediction

best_cat = grid.best_estimator_
y_pred_cat = best_cat.predict(X_test)

# 6. Evaluation

print("\n==== CATBOOST RESULTS ====")
print("R:", r2_score(y_test, y_pred_cat))
print("MAE:", mean_absolute_error(y_test, y_pred_cat))
print("MSE:", mean_squared_error(y_test, y_pred_cat))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_cat)))
# 7.visualization
plt.figure(figsize=(12,6))

plt.plot(sorted(y_test.values), label="Actual (sorted)", linewidth=2)
plt.plot(sorted(y_pred_cat), label="Predicted (sorted)", linewidth=2)

plt.title("Actual vs Predicted (Sorted Line Plot)")
plt.xlabel("Sample Index")
plt.ylabel("Power Generated")
plt.legend()
plt.show()
```

# GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score

# ------------------------------------------------------
# Define the model
# ------------------------------------------------------
gbr = GradientBoostingRegressor(random_state=42)

# Define hyperparameter search space

param_grid = {
    "n_estimators": [100, 200, 300, 400],
    "learning_rate": [0.01, 0.05, 0.1, 0.2],
    "max_depth": [1, 2, 3, 4],
    "loss": ["squared_error", "absolute_error"]
}


# Grid Search with Cross-Validation
grid_search = GridSearchCV(
    estimator=gbr,
    param_grid=param_grid,
    cv=5,                      # 5-fold Cross-validation
    scoring="r2",
    n_jobs=-1,                 # Use all CPU cores
    verbose=2
)

# Fit only on the training data
grid_search.fit(X_train, y_train)


# Best parameters found

print("Best Hyperparameters:", grid_search.best_params_)
print("Best CV R Score:", grid_search.best_score_)

# Evaluate on test data
best_model = grid_search.best_estimator_

y_pred_grad = best_model.predict(X_test)
test_r2 = r2_score(y_test, y_pred_grad)
test_mae = mean_absolute_error(y_test, y_pred_grad)
test_mse = mean_squared_error(y_test, y_pred_grad)
test_rmse = mean_squared_error(y_test, y_pred_grad) ** 0.5

print("\nTest MSE:", test_mse)
print("Test RMSE:", test_rmse)
print("Test MAE:", test_mae)
print("Test R Score:", test_r2)

#visualization
plt.figure(figsize=(12,6))

plt.plot(sorted(y_test.values), label="Actual (sorted)", linewidth=2)
plt.plot(sorted(y_pred_grad), label="Predicted (sorted)", linewidth=2)

plt.title("Actual vs Predicted (Sorted Line Plot)")
plt.xlabel("Sample Index")
plt.ylabel("Power Generated")
plt.legend()
plt.show()
```
